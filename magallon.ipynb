```python
# ==========================================================
# 🔮 ANALISIS DE TERMINACIONES DE PREMIOS POR CLIMA
# ==========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================================
# 📥 1. Cargar dataset
# ==========================================================
# Asegúrate de subir tu archivo CSV antes de ejecutar esta celda
df2 = pd.read_csv("Data_Final_exploracion_preparada.csv")

# ==========================================================
# 🕓 2. Convertir fechas
# ==========================================================
for col in ["datetime", "fecha"]:
    if col in df2.columns:
        df2[col] = pd.to_datetime(df2[col], errors="coerce")

# ==========================================================
# 🎯 3. Extraer las dos últimas cifras de los premios
# ==========================================================
df2["ult_1ro"] = df2["1ro"] % 100
df2["ult_2do"] = df2["2do"] % 100
df2["ult_3ro"] = df2["3ro"] % 100

# ==========================================================
# 🌡️ 4. Clasificar el clima
# ==========================================================
# Definición de categorías de clima (ajustables según tus datos)
def clasificar_clima(row):
    if row["lluvia_mesactual"] > 100:
        return "Lluvioso"
    elif row["temp_promedio"] > 28:
        return "Caluroso"
    elif 22 <= row["temp_promedio"] <= 28:
        return "Templado"
    else:
        return "Fresco"

df2["tipo_clima"] = df2.apply(clasificar_clima, axis=1)

# ==========================================================
# 📊 5. Calcular frecuencia de terminaciones por tipo de clima
# ==========================================================
def frecuencias_por_clima(df, col_premio):
    """Cuenta la frecuencia de terminaciones (00-99) por tipo de clima."""
    conteo = (
        df.groupby(["tipo_clima", col_premio])
          .size()
          .reset_index(name="frecuencia")
    )
    return conteo

freq_1ro = frecuencias_por_clima(df2, "ult_1ro")
freq_2do = frecuencias_por_clima(df2, "ult_2do")
freq_3ro = frecuencias_por_clima(df2, "ult_3ro")

# ==========================================================
# 🏆 6. Mostrar las terminaciones más comunes por clima
# ==========================================================
print("🔝 Top 5 terminaciones más comunes por tipo de clima:\n")

for premio, freq_df in zip(["1ro", "2do", "3ro"], [freq_1ro, freq_2do, freq_3ro]):
    print(f"\n🏅 Premio {premio}:")
    top = (
        freq_df.sort_values("frecuencia", ascending=False)
               .groupby("tipo_clima")
               .head(5)
    )
    print(top)

# ==========================================================
# 🎨 7. Visualizar las frecuencias por clima
# ==========================================================
plt.figure(figsize=(14, 8))
sns.boxplot(data=df2, x="tipo_clima", y="ult_1ro", palette="coolwarm")
plt.title("Distribución de las terminaciones del 1er premio por tipo de clima")
plt.ylabel("Últimas dos cifras del 1er premio")
plt.xlabel("Tipo de clima")
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(14, 8))
sns.countplot(data=df2, x="ult_1ro", hue="tipo_clima", palette="coolwarm")
plt.title("Frecuencia de terminaciones del 1er premio según el clima")
plt.xlabel("Terminación (00–99)")
plt.ylabel("Frecuencia")
plt.legend(title="Clima")
plt.xticks(rotation=90)
plt.show()

# ==========================================================
# 🧠 8. Analizar promedios por clima
# ==========================================================
promedios = (
    df2.groupby("tipo_clima")[["ult_1ro", "ult_2do", "ult_3ro"]]
        .mean()
        .sort_values("ult_1ro", ascending=False)
)

print("\n📊 Promedio de las dos últimas cifras de cada premio por tipo de clima:\n")
print(promedios)
```

```python
# ==========================================================
# 🧠 MODELO PROBABILÍSTICO DE TERMINACIONES SEGÚN EL CLIMA
# ==========================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================================
# 📥 1. Cargar dataset
# ==========================================================
df2 = pd.read_csv("Data_Final_exploracion_preparada.csv")

# ==========================================================
# 🕓 2. Transformar fechas
# ==========================================================
for col in ["datetime", "fecha"]:
    if col in df2.columns:
        df2[col] = pd.to_datetime(df2[col], errors="coerce")

# ==========================================================
# 🎯 3. Extraer las dos últimas cifras de los premios
# ==========================================================
df2["ult_1ro"] = df2["1ro"] % 100
df2["ult_2do"] = df2["2do"] % 100
df2["ult_3ro"] = df2["3ro"] % 100

# ==========================================================
# 🌡️ 4. Variables climáticas relevantes
# ==========================================================
variables_clima = ["temp_promedio", "lluvia_mesactual", "lluvia_acumuladaactual"]
df_clima = df2.dropna(subset=variables_clima)

# Escalar (normalizar) variables climáticas
scaler = StandardScaler()
scaled_clima = scaler.fit_transform(df_clima[variables_clima])

# Guardar matriz escalada
df_clima_scaled = df_clima.copy()
df_clima_scaled[variables_clima] = scaled_clima

# ==========================================================
# 🧩 5. Función de predicción probabilística
# ==========================================================
def predecir_terminacion_probabilistica(temp, lluvia_mes, lluvia_acum, premio="1ro", k=50):
    """
    Predice la terminación más probable según condiciones climáticas.
    Usa distancia euclidiana en el espacio normalizado de clima.
    """
    # Escalar las condiciones actuales al mismo espacio
    condiciones_scaled = scaler.transform([[temp, lluvia_mes, lluvia_acum]])

    # Calcular distancias entre condiciones históricas y actuales
    distancias = cdist(df_clima_scaled[variables_clima], condiciones_scaled)
    df_clima_scaled["distancia"] = distancias

    # Tomar los k días más similares (menor distancia)
    vecinos = df_clima_scaled.nsmallest(k, "distancia")

    # Calcular la probabilidad de cada terminación en esos vecinos
    col_premio = f"ult_{premio}"
    probabilidades = vecinos[col_premio].value_counts(normalize=True) * 100

    # Mostrar el top 5 de terminaciones más probables
    print(f"\n🔮 Predicción para condiciones: Temp={temp}°C, Lluvia mes={lluvia_mes}mm, Lluvia acumulada={lluvia_acum}mm")
    print(f"Premio analizado: {premio}\n")
    print("🎯 Terminaciones más probables:")
    print(probabilidades.head(10).round(2))

    # Visualizar distribución
    plt.figure(figsize=(10,5))
    sns.barplot(x=probabilidades.head(10).index, y=probabilidades.head(10).values, palette="viridis")
    plt.title(f"Top 10 terminaciones más probables para {premio} bajo condiciones similares")
    plt.xlabel("Terminación (00–99)")
    plt.ylabel("Probabilidad (%)")
    plt.show()

    # Devolver la terminación más probable
    return probabilidades.idxmax()

# ==========================================================
# 📈 6. Ejemplo de predicciones
# ==========================================================
# Puedes cambiar estos valores según tus escenarios
escenarios = [
    {"temp": 30, "lluvia_mes": 120, "lluvia_acum": 500, "desc": "Día caluroso y lluvioso"},
    {"temp": 25, "lluvia_mes": 60,  "lluvia_acum": 200, "desc": "Día templado con lluvias ligeras"},
    {"temp": 20, "lluvia_mes": 10,  "lluvia_acum": 50,  "desc": "Día fresco y seco"}
]

for e in escenarios:
    print(f"\n🌤️ Escenario: {e['desc']}")
    pred = predecir_terminacion_probabilistica(e["temp"], e["lluvia_mes"], e["lluvia_acum"], premio="1ro", k=50)
    print(f"👉 Terminación más probable: {pred}\n{'-'*60}")
```

```python
# ==========================================================
# 🔮 ANALISIS DE TERMINACIONES DE PREMIOS POR CLIMA Y MES
# ==========================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================================
# 📥 1. Cargar dataset
# ==========================================================
df2 = pd.read_csv("Data_Final_exploracion_preparada.csv")

# ==========================================================
# 🕓 2. Convertir fechas
# ==========================================================
for col in ["datetime", "fecha"]:
    if col in df2.columns:
        df2[col] = pd.to_datetime(df2[col], errors="coerce")

# Crear columna de mes (numérico y texto)
df2["mes"] = df2["fecha"].dt.month
df2["mes_nombre"] = df2["fecha"].dt.strftime("%B")

# ==========================================================
# 🎯 3. Extraer las dos últimas cifras de los premios
# ==========================================================
df2["ult_1ro"] = df2["1ro"] % 100
df2["ult_2do"] = df2["2do"] % 100
df2["ult_3ro"] = df2["3ro"] % 100

# ==========================================================
# 🌡️ 4. Clasificar el clima
# ==========================================================
def clasificar_clima(row):
    if row["lluvia_mesactual"] > 100:
        return "Lluvioso"
    elif row["temp_promedio"] > 28:
        return "Caluroso"
    elif 22 <= row["temp_promedio"] <= 28:
        return "Templado"
    else:
        return "Fresco"

df2["tipo_clima"] = df2.apply(clasificar_clima, axis=1)

# ==========================================================
# 📊 5. Frecuencia de terminaciones agrupadas por mes y clima
# ==========================================================
def frecuencia_mes_clima(df, col_premio):
    return (
        df.groupby(["mes_nombre", "tipo_clima", col_premio])
          .size()
          .reset_index(name="frecuencia")
    )

freq_1ro = frecuencia_mes_clima(df2, "ult_1ro")
freq_2do = frecuencia_mes_clima(df2, "ult_2do")
freq_3ro = frecuencia_mes_clima(df2, "ult_3ro")

# ==========================================================
# 🏆 6. Top de terminaciones más comunes por mes y clima
# ==========================================================
def top_mes_clima(freq_df, premio):
    top = (
        freq_df.sort_values("frecuencia", ascending=False)
               .groupby(["mes_nombre", "tipo_clima"])
               .head(1)
               .reset_index(drop=True)
    )
    top["premio"] = premio
    return top[["premio", "mes_nombre", "tipo_clima", f"ult_{premio.replace('ro','')}", "frecuencia"]]

top_1ro = top_mes_clima(freq_1ro, "1ro")
top_2do = top_mes_clima(freq_2do, "2do")
top_3ro = top_mes_clima(freq_3ro, "3ro")

top_todos = pd.concat([top_1ro, top_2do, top_3ro])
print("🏅 Terminaciones más frecuentes por MES y CLIMA:\n")
print(top_todos.sort_values(["mes_nombre", "premio"]))

# ==========================================================
# 🎨 7. Visualizar tendencias por mes
# ==========================================================
plt.figure(figsize=(14,6))
sns.lineplot(data=df2, x="mes_nombre", y="ult_1ro", hue="tipo_clima", marker="o")
plt.title("Tendencia mensual de terminaciones del 1er premio por clima")
plt.ylabel("Últimas dos cifras (1er premio)")
plt.xlabel("Mes")
plt.legend(title="Tipo de clima")
plt.grid(alpha=0.3)
plt.show()

# ==========================================================
# 📈 8. Promedios de terminaciones por mes y clima
# ==========================================================
promedios_mes = (
    df2.groupby(["mes_nombre", "tipo_clima"])[["ult_1ro", "ult_2do", "ult_3ro"]]
       .mean()
       .round(2)
       .reset_index()
)

print("\n📊 Promedio de las dos últimas cifras de premios por mes y clima:\n")
print(promedios_mes)

# ==========================================================
# 🧠 9. Conclusión resumida
# ==========================================================
clima_promedio = (
    promedios_mes.groupby("tipo_clima")[["ult_1ro", "ult_2do", "ult_3ro"]]
    .mean()
    .round(2)
    .sort_values("ult_1ro", ascending=False)
)

print("\n🏆 Climas con terminaciones promedio más altas en general:")
print(clima_promedio)
```

```python
# ==========================================================
# 🔮 PREDICCIÓN DE TERMINACIONES PARA OCTUBRE (4 SORTEOS)
# ==========================================================

import pandas as pd
import numpy as np

# ------------------------------------------
# 1️⃣ Filtrar registros de octubre
# ------------------------------------------
df_oct = df2[df2["mes_nombre"] == "October"].copy()

# ------------------------------------------
# 2️⃣ Clasificar días como "Miercolito" o "Dominical"
# ------------------------------------------
df_oct["dia_semana"] = df_oct["fecha"].dt.day_name()

def tipo_sorteo(dia):
    if dia == "Wednesday":
        return "Miercolito"
    elif dia == "Sunday":
        return "Dominical"
    else:
        return "Otro"

df_oct["tipo_sorteo"] = df_oct["dia_semana"].apply(tipo_sorteo)

# Filtrar solo miercolitos y dominicales
df_oct = df_oct[df_oct["tipo_sorteo"].isin(["Miercolito", "Dominical"])]

# ------------------------------------------
# 3️⃣ Calcular probabilidades por tipo de sorteo
# ------------------------------------------
def calcular_probabilidades(df, columna):
    conteo = df[columna].value_counts().sort_index()
    total = conteo.sum()
    prob = (conteo / total * 100).round(2)
    resultado = pd.DataFrame({
        "Terminacion": conteo.index,
        "Frecuencia": conteo.values,
        "Probabilidad_%": prob.values
    }).sort_values("Probabilidad_%", ascending=False)
    return resultado

def prediccion_por_tipo(df, tipo):
    subset = df[df["tipo_sorteo"] == tipo]
    p1 = calcular_probabilidades(subset, "ult_1ro")
    p2 = calcular_probabilidades(subset, "ult_2do")
    p3 = calcular_probabilidades(subset, "ult_3ro")
    return p1, p2, p3

# Obtener resultados
prob_miercolito = prediccion_por_tipo(df_oct, "Miercolito")
prob_dominical = prediccion_por_tipo(df_oct, "Dominical")

# ------------------------------------------
# 4️⃣ Mostrar predicciones
# ------------------------------------------
def mostrar_top(p1, p2, p3, titulo):
    print(f"\n🎯 PREDICCIÓN PARA {titulo.upper()} (TOP 5 más probables)\n")
    print("🏆 1er Premio:")
    print(p1[0].head(5).to_string(index=False))
    print("\n🥈 2do Premio:")
    print(p2[0].head(5).to_string(index=False))
    print("\n🥉 3er Premio:")
    print(p3[0].head(5).to_string(index=False))

# Mostrar
mostrar_top(prob_miercolito, prob_miercolito, prob_miercolito, "Miercolito")
mostrar_top(prob_dominical, prob_dominical, prob_dominical, "Dominical")

# ------------------------------------------
# 5️⃣ Crear predicción combinada para 4 sorteos
# ------------------------------------------
def top_promedio(p1, p2, p3, n=5):
    comb = pd.concat([
        p1[0][["Terminacion", "Probabilidad_%"]],
        p2[0][["Terminacion", "Probabilidad_%"]],
        p3[0][["Terminacion", "Probabilidad_%"]],
    ])
    return (comb.groupby("Terminacion")["Probabilidad_%"]
            .mean()
            .sort_values(ascending=False)
            .head(n))

print("\n🔮 Predicción combinada (probabilidad promedio global):")
pred_global_miercolito = top_promedio(prob_miercolito, prob_miercolito, prob_miercolito)
pred_global_dominical = top_promedio(prob_dominical, prob_dominical, prob_dominical)

print("\n🧾 Miercolitos:")
print(pred_global_miercolito.to_string())

print("\n🧾 Dominicales:")
print(pred_global_dominical.to_string())

# ------------------------------------------
# 6️⃣ Generar 4 sorteos simulados
# ------------------------------------------
def simular_sorteo(prob_df):
    return np.random.choice(prob_df.index, p=(prob_df / prob_df.sum()))

print("\n🎲 SORTEOS SIMULADOS (predicción basada en probabilidad):\n")
for i in range(1, 3):
    num = simular_sorteo(pred_global_miercolito)
    print(f"➡️ Miercolito {i}: terminación probable {int(num):02d}")

for i in range(1, 3):
    num = simular_sorteo(pred_global_dominical)
    print(f"➡️ Dominical {i}: terminación probable {int(num):02d}")
```

```python
# ==========================================================
# 🔮 PREDICCIÓN DE DECENAS PARA OCTUBRE (4 SORTEOS)
# ==========================================================

import pandas as pd
import numpy as np

# ------------------------------------------
# 1️⃣ Filtrar registros de octubre
# ------------------------------------------
df_oct = df2[df2["mes_nombre"] == "October"].copy()

# ------------------------------------------
# 2️⃣ Clasificar días como "Miercolito" o "Dominical"
# ------------------------------------------
df_oct["dia_semana"] = df_oct["fecha"].dt.day_name()

def tipo_sorteo(dia):
    if dia == "Wednesday":
        return "Miercolito"
    elif dia == "Sunday":
        return "Dominical"
    else:
        return "Otro"

df_oct["tipo_sorteo"] = df_oct["dia_semana"].apply(tipo_sorteo)
df_oct = df_oct[df_oct["tipo_sorteo"].isin(["Miercolito", "Dominical"])]

# ------------------------------------------
# 3️⃣ Extraer la decena (ejemplo: 47 → 40)
# ------------------------------------------
df_oct["decena_1ro"] = (df_oct["ult_1ro"] // 10) * 10
df_oct["decena_2do"] = (df_oct["ult_2do"] // 10) * 10
df_oct["decena_3ro"] = (df_oct["ult_3ro"] // 10) * 10

# ------------------------------------------
# 4️⃣ Calcular probabilidades por tipo de sorteo
# ------------------------------------------
def calcular_probabilidades_decenas(df, columna):
    conteo = df[columna].value_counts().sort_index()
    total = conteo.sum()
    prob = (conteo / total * 100).round(2)
    resultado = pd.DataFrame({
        "Decena": conteo.index,
        "Frecuencia": conteo.values,
        "Probabilidad_%": prob.values
    }).sort_values("Probabilidad_%", ascending=False)
    return resultado

def prediccion_decenas_por_tipo(df, tipo):
    subset = df[df["tipo_sorteo"] == tipo]
    p1 = calcular_probabilidades_decenas(subset, "decena_1ro")
    p2 = calcular_probabilidades_decenas(subset, "decena_2do")
    p3 = calcular_probabilidades_decenas(subset, "decena_3ro")
    return p1, p2, p3

# Calcular probabilidades por tipo de sorteo
prob_dec_miercolito = prediccion_decenas_por_tipo(df_oct, "Miercolito")
prob_dec_dominical = prediccion_decenas_por_tipo(df_oct, "Dominical")

# ------------------------------------------
# 5️⃣ Mostrar resultados
# ------------------------------------------
def mostrar_top_decenas(prob, titulo):
    print(f"\n📊 PREDICCIÓN DE DECENAS PARA {titulo.upper()} DE OCTUBRE (TOP 5)\n")
    print("🏆 1er Premio:")
    print(prob[0].head(5).to_string(index=False))
    print("\n🥈 2do Premio:")
    print(prob[1].head(5).to_string(index=False))
    print("\n🥉 3er Premio:")
    print(prob[2].head(5).to_string(index=False))

mostrar_top_decenas(prob_dec_miercolito, "Miercolito")
mostrar_top_decenas(prob_dec_dominical, "Dominical")

# ------------------------------------------
# 6️⃣ Promedio global de probabilidades por decena
# ------------------------------------------
def top_promedio_decenas(p1, p2, p3, n=5):
    comb = pd.concat([
        p1[["Decena", "Probabilidad_%"]],
        p2[["Decena", "Probabilidad_%"]],
        p3[["Decena", "Probabilidad_%"]],
    ])
    return (comb.groupby("Decena")["Probabilidad_%"]
            .mean()
            .sort_values(ascending=False)
            .head(n))

print("\n🔮 PREDICCIÓN COMBINADA POR DECENAS (probabilidad promedio global)\n")
pred_dec_miercolito = top_promedio_decenas(prob_dec_miercolito[0], prob_dec_miercolito[1], prob_dec_miercolito[2])
pred_dec_dominical = top_promedio_decenas(prob_dec_dominical[0], prob_dec_dominical[1], prob_dec_dominical[2])

print("🧾 Miercolitos:")
print(pred_dec_miercolito.to_string())

print("\n🧾 Dominicales:")
print(pred_dec_dominical.to_string())

# ------------------------------------------
# 7️⃣ Simulación de 4 sorteos (2 miercolitos, 2 dominicales)
# ------------------------------------------
def simular_decena(prob_df):
    return np.random.choice(prob_df.index, p=(prob_df / prob_df.sum()))

print("\n🎲 SORTEOS SIMULADOS POR DECENA PARA OCTUBRE:\n")
for i in range(1, 3):
    num = simular_decena(pred_dec_miercolito)
    print(f"➡️ Miercolito {i}: decena probable {int(num):02d}s")

for i in range(1, 3):
    num = simular_decena(pred_dec_dominical)
    print(f"➡️ Dominical {i}: decena probable {int(num):02d}s")
```

```python
# ==========================================================
# 🔮 EXTRAPOLACIÓN DE LAS DOS ÚLTIMAS CIFRAS PARA OCTUBRE
# ==========================================================

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# --------------------------------------------------------
# 1️⃣ Filtrar los datos de OCTUBRE
# --------------------------------------------------------
df_oct = df2[df2["mes_nombre"] == "October"].copy()
df_oct = df_oct.dropna(subset=["fecha"])
df_oct["anio"] = df_oct["fecha"].dt.year

# --------------------------------------------------------
# 2️⃣ Asegurar que existen las columnas de últimas cifras
# --------------------------------------------------------
if not {"ult_1ro", "ult_2do", "ult_3ro"}.issubset(df_oct.columns):
    df_oct["ult_1ro"] = df_oct["1ro"] % 100
    df_oct["ult_2do"] = df_oct["2do"] % 100
    df_oct["ult_3ro"] = df_oct["3ro"] % 100

# --------------------------------------------------------
# 3️⃣ Unificar los tres premios en un solo conjunto
# --------------------------------------------------------
df_long = pd.melt(df_oct,
                  id_vars=["anio"],
                  value_vars=["ult_1ro", "ult_2do", "ult_3ro"],
                  var_name="premio",
                  value_name="terminacion")

# --------------------------------------------------------
# 4️⃣ Calcular frecuencia por año y terminación
# --------------------------------------------------------
freq_anual = (
    df_long.groupby(["anio", "terminacion"])
    .size()
    .reset_index(name="frecuencia")
)

# --------------------------------------------------------
# 5️⃣ Extrapolación por regresión lineal (año → frecuencia)
# --------------------------------------------------------
predicciones = []
años = freq_anual["anio"].unique().reshape(-1, 1)
año_futuro = [[años.max() + 1]]

for term in sorted(freq_anual["terminacion"].unique()):
    subset = freq_anual[freq_anual["terminacion"] == term]
    if len(subset) > 2:  # Evitar ajustar sobre pocas observaciones
        X = subset["anio"].values.reshape(-1, 1)
        y = subset["frecuencia"].values
        modelo = LinearRegression()
        modelo.fit(X, y)
        y_pred = modelo.predict(año_futuro)[0]
        predicciones.append((term, y_pred))

# --------------------------------------------------------
# 6️⃣ Resultados extrapolados
# --------------------------------------------------------
pred_df = pd.DataFrame(predicciones, columns=["Terminación", "Frecuencia_Predicha"])
pred_df = pred_df.sort_values("Frecuencia_Predicha", ascending=False)

print("📈 EXTRAPOLACIÓN DE TERMINACIONES (DOS ÚLTIMAS CIFRAS) PARA OCTUBRE:\n")
print(pred_df.head(15).to_string(index=False))

# --------------------------------------------------------
# 7️⃣ Visualización de las terminaciones con tendencia positiva
# --------------------------------------------------------
top_terms = pred_df.head(5)["Terminación"].tolist()

plt.figure(figsize=(10,6))
for term in top_terms:
    sub = freq_anual[freq_anual["terminacion"] == term]
    plt.plot(sub["anio"], sub["frecuencia"], marker="o", label=f"Terminación {int(term):02d}")

plt.title("📊 Tendencia histórica de las terminaciones más fuertes en OCTUBRE")
plt.xlabel("Año")
plt.ylabel("Frecuencia de aparición")
plt.legend(title="Terminación")
plt.grid(alpha=0.3)
plt.show()
```
